# NUMERAL 4

diagrama de dispersión para verificar sí la edad tiene influencia con los ingresos 

```python
import seaborn as sns
sns.set_theme(color_codes=True)
datos=data[["edad","ingresos"]]
df=pd.DataFrame(datos)

sns.regplot(x="edad", y="ingresos", data=df);
```

La edad si influye en los ingresos
